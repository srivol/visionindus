```python
%matplotlib inline
# chargement des modules python
import matplotlib.pyplot as plt
import numpy as np
import skimage
from skimage import data
```

# Chargement d'une image

```python
camera = data.camera()
plt.imshow(camera, cmap=plt.cm.gray)
plt.show()
```

# Chargement d'une image depuis le disque

```python
cat = skimage.data.chelsea()
cat = cat[:,:,1] # image à niveaux de gris, canal vert
plt.imshow(cat, cmap=plt.cm.gray)
plt.show()
```

# Affichage de l'histogramme de l'image n&b

```python
y,x=np.histogram(cat,256,range=(0,255))
plt.plot(x[:-1],y)
plt.show()
```

# Ajout d'une constante

Objectif:
- Ajouter/soustraire une valeur constante à l'image (à tous ses pixels)
- Vérifier l'impact sur l'histogramme de l'image

```python
addition=cat+50
plt.imshow(cat, cmap=plt.cm.gray, vmin=0, vmax=255)
plt.show()
plt.imshow(addition, cmap=plt.cm.gray, vmin=0, vmax=255)
plt.show()
yadd,xadd=np.histogram(addition,256,range=(0,255))
plt.plot(xadd[:-1],yadd)
plt.show()
```
